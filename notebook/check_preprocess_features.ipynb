```python
from sklearn.preprocessing import LabelEncoder
```

```python
def list_label_encoder(train, test, columns: list):
    le = LabelEncoder()
    train_description_val_session_title = train[columns].apply(lambda x: str([i for i in x]), axis=1)
    test_description_val_session_title = test[columns].apply(lambda x: str([i for i in x]), axis=1)
    le.fit(train_description_val_session_title)
    le_dict = dict(zip(le.classes_, le.transform(le.classes_)))
    label_name = '_'.join(columns)
    train[f'label_{label_name}'] = train_description_val_session_title.apply(lambda x: le_dict.get(x, -1))
    test[f"label_{label_name}"] = test_description_val_session_title.apply(lambda x: le_dict.get(x, -1))
    return train, test

def frequency_encoder(train, test, columns):
    label_name = '_'.join(columns)
    freq = train[label_name].value_counts()
    train[f'count_{label_name}'] = train[label_name].map(freq)
    test[f'count_{label_name}'] = test[label_name].map(freq)
    return train, test
```

```python
def count_accuracy_label_encoder(x):
    if x == 0:
        y = 0
    elif x > 0 and x <= 0.34:
        y = 1
    elif x > 0.34 and x <= 0.67:
        y = 2
    elif x > 0.67 and x < 1:
        y = 3
    elif x == 1:
        y = 4
    return y
```

```python
train['count_accuracy_label'] = train.count_accuracy.apply(lambda x: count_accuracy_label_encoder(x))
test['count_accuracy_label'] = test.count_accuracy.apply(lambda x: count_accuracy_label_encoder(x))
```

```python
train, test = list_label_encoder(train, test, ['session_title', 'count_accuracy_label'])
train, test = frequency_encoder(train, test, ['session_title', 'count_accuracy_label'])
```
